```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
                                             #PLOTLY ASSIGNMENT#
#Q1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three dimensional space*

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
df
fig = go.Figure()
fig.add_traces(go.Scatter3d(x = df['X'],y = df['Y'],z= df['Z'],mode = 'markers'))
```

```python
#Q2.. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories#

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
sns.catplot(x = 'Grade',y = 'Score', data = df, kind = 'violin')

```

```python
#Q3 and Q2.1 are same question :  Using the sales data, generate a heatmap to visualize the variation in sales across different months and days#

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 8))
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum')#to get days as rows and months as columns

sns.heatmap(sales_pivot, annot=True, fmt=".0f", cmap='YlGnBu', cbar_kws={'label': 'Sales'})
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()

```

```python
#Q4. Using the given x and y data, generate a 3D surface plot to visualize the function : z = sin(√x^2 + y^2 )   #

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure()
fig.add_traces(go.Scatter3d(x = df['X'],y = df['Y'],z = df['Z'] , mode ='markers'))

```

```python
                                                                              #BOKEH ASSIGNMENT#
#Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

#pip install bokeh

import numpy as np
import pandas as pd
import bokeh.io
import bokeh.application
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook


output_notebook()
x = np.linspace(0,10,100)
y = np.sin(x)

p = figure(title = 'sien wave', x_axis_label = 'x',y_axis_label = 'sin(x)',width = 900, height = 450)
p.line(x,y,legend_label="Sine Wave",line_width=2)
show(p)
```

```python
#Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns#

np.random.seed(45)
n = 120
x = np.random.rand(n) *10
y = np.random.rand(n) *10
sizes = np.random.rand(n) *25+10
colors = np.random.rand(n, 4)

df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors.tolist()})
output_file("Scatter.html")
p = figure(title ='Scatter plot with variable sizes and colors', x_axis_label='x', y_axis_label='y')
p.circle(x=df['x'],y=df['y'],size = df['sizes'],color =[tuple(c) for c in df['colors']],alpha=0.6)
show(p)
```

```python
#Q3.  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.#

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

df = pd.DataFrame({'x':fruits,'y':counts})
output_file('bar.html')
p = figure(title ='counts of different fruits',x_range=df['x'],x_axis_label = 'Fruits', y_axis_label = 'Counts' )
p.vbar(x=df['x'],top=df['y'],color = 'brown',width=0.5, alpha=0.6)
show(p)
```

```python
#Q4 Create a Bokeh histogram to visualize the distribution of the given data#

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

output_file('histogram.html')
p= figure(title = 'Histogram to Visualize The Distribution', x_axis_label = 'value', y_axis_label = 'Frequency')
p.quad(top=hist,bottom =0,left=edges[:-1],right=edges[1:],fill_color = 'brown',line_color = 'DarkBrown',alpha = 0.6)
show(p)
```

```python
#Q5. Create a Bokeh heatmap using the provided dataset.

from bokeh.models import LinearColorMapper
from bokeh.palettes import YlGnBu

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title ='Bokeh heatmap',x_axis_label = "X",y_axis_label = "Y",x_range=(0, 1), y_range=(0, 1),tools='')
color_mapper = LinearColorMapper(palette=YlGnBu[9], low=data_heatmap.min(), high=data_heatmap.max())
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1,color_mapper=color_mapper)

show(p)
```
